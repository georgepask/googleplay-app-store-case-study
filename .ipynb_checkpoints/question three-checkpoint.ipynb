For your category, how many genres are represented? What are their counts? Show in a bar chart.

```python
# Examine App categories
google_play_df.Category.unique()
```

```python
# Pull GAME category from dataframe
df1 = google_play_df.loc[google_play_df['Category'] == 'GAME']
df1
```

```python
# Examine Genres within GAME, notice some have duplicate descriptors
df2 = df1.groupby('Genres')['App'].nunique().reset_index()
df2
```

```python
# Split genres with duplicate descriptors
df1['genres_lists']  = df1['Genres'].str.split(";")
df1.head(2)
```

```python
# Place all genres into a single list
total_list = []
for lst in df1['genres_lists'].values:
    total_list.extend(lst)
total_list[:10]
```

```python
# Count occurances of each genre a
from collections import Counter
counter = Counter(total_list)
# Sort by descending values
srt = sorted(counter.items(), key=lambda x: x[1], reverse=True)
# Create plot axis of (Genre, Counts)
x_val= list(map(lambda x: x[0], srt))
y_val = list(map(lambda x: x[1], srt))
srt
```

```python
# Total unique genres
len(srt)
```

```python
# Total games
tot = sum(y_val) 
tot
```

```python
# Percentage occupied by top 3 genres
round(368/tot, 2), round(220/tot, 2), round(98/tot, 2)
```

```python
# Please show this with a bar chart. 
plt.figure(figsize=(16, 6))
g = sns.barplot(x=x_val, y=y_val)
plt.title('Genres of Game Apps on Google Play Store')
plt.xticks(rotation=90)
plt.ylabel('Number of Apps')
plt.xlabel('Genres')
plt.show()
```
